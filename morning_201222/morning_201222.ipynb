```python
#step1. 엑셀파일 만든느 함수 연습_5
import openpyxl
def write_excel_template(filename, sheetname, listdata):
    excel_file = openpyxl.Workbook()
    excel_sheet = excel_file.active
    excel_sheet.column_dimensions['A'].width = 100
    
    if sheetname != '':
        excel_sheet.title = sheetname
        
    for item in listdata:
        excel_sheet.append(item)
    excel_file.save(filename)
    excel_file.close()
```

```python
#step2. 여러 페이지 읽어오기1
import requests
from bs4 import BeautifulSoup
product_lists = list()

res_1 = requests.get('https://davelee-fun.github.io/')
soup_1 = BeautifulSoup(res_1.content, 'html.parser')
data_top = soup_1.select('div.col-12.col-md-12.col-lg-7')

for data in data_top:
    product_name_top = data.select_one('h4.card-text')
    product_data_top = data.select_one('span.post-date')
    product_info_top = [product_name_top.get_text(), product_data_top.get_text()]
    product_lists.append(product_info_top)
for page_num in range(10):
    if page_num ==0:
        res = requests.get('https://davelee-fun.github.io/')
    else:
        res = requests.get('https://davelee-fun.github.io/page' + str(page_num +1))
    soup = BeautifulSoup(res.content, 'html.parser')
    data = soup.select('div.col-lg-4.col-md-6.mb-30px.card-group')
    for item in data:
        product_name = item.select_one('h4.card-text')
        product_date = item.select_one('span.post-date')
        product_info = [product_name.get_text(), product_date.get_text()]
        product_lists.append(product_info)
for index, product_list in enumerate(product_lists):
    print(index, product_list)
```
