```python
import requests
import openpyxl
from bs4 import BeautifulSoup
work_book = openpyxl.Workbook()
work_sheet = work_book.active
work_sheet.append(['상품순위','상품명','가격','소싱처','링크'])
work_sheet.title = '지마켓_100'

work_sheet.column_dimensions['A'].width = 15
work_sheet.column_dimensions['B'].width = 50
work_sheet.column_dimensions['C'].width = 20
work_sheet.column_dimensions['D'].width = 20
work_sheet.column_dimensions['E'].width = 50

product_info = list()
res = requests.get('http://corners.gmarket.co.kr/Bestsellers?viewType=G&groupCode=G07')
soup = BeautifulSoup(res.content, 'html.parser')

best_lists = soup.select('div.best-list')
best_list = best_lists[1]
card = best_list.select('ul>li')

for index, item in enumerate(card):
    product = item.select_one('a.itemname')
    price = item.select_one('div.s-price > strong')
    res_1 = requests.get(product['href'])
    soup_1 = BeautifulSoup(res_1.content, 'html.parser')
    company_name = soup_1.select_one('div.item-topinfo a.shopurl strong')
    print(index+1, product.get_text(), price.get_text(), company_name.get_text(), product['href'])
    work_sheet.append([index+1, product.get_text(), price.get_text(), company_name.get_text(), product['href']])
    work_sheet.cell(row = index +2, column=5).hyperlink = product['href']
cell_A1 = work_sheet['A1']
cell_A1.alignment = openpyxl.styles.Alignment(horizontal='center')
cell_A1.font = openpyxl.styles.Font(color='48a999')

cell_B1 = work_sheet['B1']
cell_B1.alignment = openpyxl.styles.Alignment(horizontal='center')
cell_B1.font = openpyxl.styles.Font(color='48a999')

cell_C1 = work_sheet['C1']
cell_C1.alignment = openpyxl.styles.Alignment(horizontal='center')
cell_C1.font = openpyxl.styles.Font(color='48a999')

cell_D1 = work_sheet['D1']
cell_D1.alignment = openpyxl.styles.Alignment(horizontal='center')
cell_D1.font = openpyxl.styles.Font(color='48a999')

cell_E1 = work_sheet['E1']
cell_E1.alignment = openpyxl.styles.Alignment(horizontal='center')
cell_E1.font = openpyxl.styles.Font(color='48a999')

work_book.save('g_market_best100.xlsx')
work_book.close()
```
