```python
import openpyxl
import re

work_book=openpyxl.load_workbook('../../regular_expression/train.xlsx')
work_sheet=work_book.active

man_work_sheet = work_book.create_sheet('남성')
solo_woman_work_sheet = work_book.create_sheet('미혼여성')
married_woman_work_sheet = work_book.create_sheet('기혼여성')
others_work_sheet=work_book.create_sheet('기타')
analysis_work_sheet=work_book.create_sheet('생존율')

man_work_sheet.column_dimensions['D'].width = 50
solo_woman_work_sheet.column_dimensions['D'].width = 50
married_woman_work_sheet.column_dimensions['D'].width = 50
others_work_sheet.column_dimensions['D'].width = 50

analysis_work_sheet.column_dimensions['A'].width = 5
analysis_work_sheet.column_dimensions['B'].width = 10
analysis_work_sheet.column_dimensions['C'].width = 10
analysis_work_sheet.column_dimensions['D'].width = 10

man_survived, man_unsurvived = 0,0
solo_woman_survived, solo_woman_unsurvived = 0,0
married_woman_survived, married_woman_unsurvived = 0,0
other_survived, other_unsurvived = 0,0
regex = re.compile(' [A-Za-z]+\.')

for each_row in work_sheet.rows: 
    data = regex.findall(each_row[3].value)
    
    if each_row[0].row == 1:
        man_work_sheet.append([item.value for item in each_row])
        solo_woman_work_sheet.append([item.value for item in each_row])
        married_woman_work_sheet.append([item.value for item in each_row])
        others_work_sheet.append([item.value for item in each_row])
    else:
        if len(data)>0:
            if data[0] == " Mr.":
                man_work_sheet.append([item.value for item in each_row])
                if each_row[1].value == 1:
                    man_survived +=1
                else:
                    man_unsurvived +=1
            elif data[0] == " Miss.":
                solo_woman_work_sheet.append([item.value for item in each_row])
                if each_row[1].value == 1:
                    solo_woman_survived +=1
                else:
                    solo_woman_unsurvived +=1
            elif data[0] == " Mrs.":
                married_woman_work_sheet.append([item.value for item in each_row])
                if each_row[1].value == 1:
                    married_woman_survived +=1
                else:
                    married_woman_unsurvived +=1
            else:
                others_work_sheet.append([item.value for item in each_row])
                if each_row[1] ==1:
                    other_survived +=1
                else:
                    other_unsurvived +=1
analysis_work_sheet.append(['','','',''])
analysis_work_sheet.append(['','분류','생존자수','사망자수','생존율'])

man_survived_rate = "%.2f%%" % (man_survived / (man_survived + man_unsurvived) * 100)
analysis_work_sheet.append(["","남성",man_survived,man_unsurvived,man_survived_rate])
solo_woman_survived_rate = "%.2f%%"%(solo_woman_survived/(solo_woman_survived+solo_woman_unsurvived)*100)
analysis_work_sheet.append(['','미혼여성',solo_woman_survived,solo_woman_unsurvived,solo_woman_survived_rate])      
married_woman_survived_rate = "%.2f%%"%(married_woman_survived/(married_woman_survived+married_woman_unsurvived)*100)
analysis_work_sheet.append(['','기혼여성',married_woman_survived,married_woman_unsurvived,married_woman_survived_rate])      
other_survived_rate = "%.2f%%"%(other_survived/(other_survived+other_unsurvived)*100)
analysis_work_sheet.append(['','기타',other_survived,other_unsurvived,other_survived_rate])

# others_survived_rate = "%.2f%%" % (others_survived / (others_survived + others_unsurvived) * 100)
# work_sheet_report.append(['', '기타', others_survived, others_unsurvived, others_survived_rate])



work_book.save('tran_analysis_0.xlsx')
work_book.close()
```
